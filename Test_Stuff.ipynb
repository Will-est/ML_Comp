```python
import pandas as pd
import numpy as np
from sklearn import datasets

data = pd.read_csv("train.csv")
```

```python
data.head()
```

**Tests Correlation**

```python
# Create a correlation matrix between categorical features and Outcome Type
categorical_features = ['Intake Type', 'Intake Condition', 'Animal Type', 'Sex upon Intake', 'Breed', 'Color']

# Create a dictionary to store the contingency tables
contingency_tables = {}

# For each categorical feature, calculate the contingency table with Outcome Type
for feature in categorical_features:
    # Create cross-tabulation (contingency table)
    contingency = pd.crosstab(data[feature], data['Outcome Type'])
    
    # Store the contingency table
    contingency_tables[feature] = contingency
    
    # Display the table with proper formatting
    print(f"\n### {feature} vs Outcome Type ###")
    print(contingency)
    
    # Calculate and display percentages (how likely each outcome is given the feature value)
    percentage = contingency.div(contingency.sum(axis=1), axis=0) * 100
    print("\nPercentage by row (%):")
    print(percentage.round(2))
```
